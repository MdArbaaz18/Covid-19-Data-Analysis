```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('owid-covid-data.csv')
df.head()
```

```python
df=df[['location','date','total_deaths']]
df.dropna(inplace=True)
```

```python
countries=['India','United States','Brazil','United Kingdom','Germany']
df=df[df['location'].isin(countries)]
```

```python
plt.figure(figsize=(10,5))
for country in countries:
  data=df[df['location']==country]
  plt.plot(data['date'],data['total_deaths'],label=country)
plt.xlabel('Date')
plt.ylabel('Total Deaths')
plt.title('COVID-19 Total Deaths Over Time')
plt.legend()
plt.show()
```

##Conclusion
->COVID-19 cases increased at different rates across countries.  
->Some countries experienced sharper growth compared to others.
